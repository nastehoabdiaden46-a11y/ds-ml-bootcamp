# Loan Approval Classification

## Objective
This notebook predicts loan approval using three classification models:

1. Logistic Regression
2. Random Forest
3. Decision Tree

The target variable is:

Approved = 1  
Rejected = 0

### Third Algorithm Research

Chosen Algorithm: Decision Tree Classifier

Reason:
Decision Tree is suitable for loan approval prediction because it creates decisions based on conditions such as credit score, income, and employment years.

Advantage:
Easy to understand and interpret.

Limitation:
Can overfit training data.

Source:
https://scikit-learn.org/stable/modules/tree.html++++++++++++++++++++++++

```python
# ==========================================
# Assignment Five Part B2
# Loan Approval Classification
# ==========================================

# Import Libraries

import pandas as pd

from sklearn.model_selection import train_test_split

from sklearn.linear_model import LogisticRegression

from sklearn.ensemble import RandomForestClassifier

from sklearn.tree import DecisionTreeClassifier

from sklearn.metrics import (
    
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    confusion_matrix

)

print("Libraries Imported Successfully")


# ==========================================
# Load Dataset
# ==========================================

df = pd.read_csv("clean_loan_dataset.csv")

print("\nDataset Loaded Successfully")

print("\nFirst 5 Rows:")

print(df.head())


# ==========================================
# Prepare Features and Target
# ==========================================

y = df["Approved"]

X = df.drop(
    
    "Approved",
    axis=1

)

print("\nFeatures Shape:")

print(X.shape)

print("\nTarget Shape:")

print(y.shape)


# ==========================================
# Split Dataset
# ==========================================

X_train, X_test, y_train, y_test = train_test_split(
    
    X,
    y,
    test_size=0.20,
    stratify=y,
    random_state=42
)

print("\nTrain Shape:")

print(X_train.shape)

print("\nTest Shape:")

print(X_test.shape)


# ==========================================
# Train Models
# ==========================================


logistic_model = LogisticRegression(
    
    max_iter=1000,
    random_state=42
)


rf_model = RandomForestClassifier(
    
    n_estimators=100,
    random_state=42
)


dt_model = DecisionTreeClassifier(
    
    random_state=42
)



logistic_model.fit(
    
    X_train,
    y_train
)

rf_model.fit(
    
    X_train,
    y_train
)

dt_model.fit(
    
    X_train,
    y_train
)


print("\nAll Models Trained Successfully")


# ==========================================
# Evaluation Function
# ==========================================

def evaluate_model(model,name):


    prediction = model.predict(
        
        X_test
    )


    accuracy = accuracy_score(
        
        y_test,
        prediction
    )


    precision = precision_score(
        
        y_test,
        prediction,
        pos_label=1
    )


    recall = recall_score(
        
        y_test,
        prediction,
        pos_label=1
    )


    f1 = f1_score(
        
        y_test,
        prediction,
        pos_label=1
    )


    print("\n")

    print(name)

    print(
        f" Accuracy : {accuracy:.3f}"
    )

    print(
        f" Precision: {precision:.3f}"
        + " (positive=Approved=1)"
    )


    print(
        f" Recall   : {recall:.3f}"
        + " (positive=Approved=1)"
    )


    print(
        f" F1-Score : {f1:.3f}"
    )


    cm = confusion_matrix(
        
        y_test,
        prediction
    )


    print("\nConfusion Matrix")

    print(cm)


# ==========================================
# Evaluate Models
# ==========================================


evaluate_model(
    
    logistic_model,
    "Logistic Regression Performance:"
)


evaluate_model(
    
    rf_model,
    "Random Forest Performance:"
)


evaluate_model(
    
    dt_model,
    "Decision Tree Performance:"
)


# ==========================================
# Sanity Check
# ==========================================

print("\n")

print("================================")

print("Sanity Check")

print("================================")


i = 0


sample = X_test.iloc[[i]]

actual = y_test.iloc[i]


logistic_prediction = logistic_model.predict(
    
    sample
)

rf_prediction = rf_model.predict(
    
    sample
)

dt_prediction = dt_model.predict(
    
    sample
)


print(
    
    "Actual Label:",
    actual
)


print(
    
    "Logistic Prediction:",
    logistic_prediction[0]
)


print(
    
    "Random Forest Prediction:",
    rf_prediction[0]
)


print(
    
    "Decision Tree Prediction:",
    dt_prediction[0]
)



print("\nNotebook Completed Successfully")
```
